```python
#filling missing values with mean or median instead of dropping the values
```

```python
df['fixed_acidity'].fillna(df['fixed_acidity'].mean(), inplace=True)
df['volatile_acidity'].fillna(df['volatile_acidity'].median(), inplace=True)
df['citric_acid'].fillna(df['citric_acid'].mean(), inplace=True)
df['residual_sugar'].fillna(df['residual_sugar'].mean(), inplace=True)
df['chlorides'].fillna(df['chlorides'].median(), inplace=True)
df['pH'].fillna(df['pH'].mean(), inplace=True)
df['sulphates'].fillna(df['sulphates'].median(), inplace=True)
```

```python
uv1=np.percentile(df.free_sulfur_dioxide,[99])[0]
uv1
df[df.free_sulfur_dioxide>uv1*2]
df[df.free_sulfur_dioxide>uv1]=uv1
df[df.free_sulfur_dioxide>uv1]

```

```python
#removal of outlier in total sufurdioxide
uv2=np.percentile(df.total_sulfur_dioxide,[99])[0]
uv2
df[df.total_sulfur_dioxide>uv2*2]
df[df.free_sulfur_dioxide>uv2]=uv2
df[df.free_sulfur_dioxide>uv2]
```
